```python
import pandas as pd
import numpy as np
```

```python
pd.read_csv('files/metric/eval_by_metrix/comments_mae.csv')
```

```python
metrix_list = ['views','comments','likes']
metric_list = ['rmse','r2','mae']
```

```python
result={}
index = None
for metrix in metrix_list:
    for metric in metric_list:
        data = pd.read_csv(f'files/metric/eval_by_metrix/{metrix}_{metric}.csv',index_col=0)
        new_row = data['mean']
        index = data.index
        result[f'{metrix}_{metric}'] = new_row.tolist()
        
```

```python
df = pd.DataFrame(data=result,index=index)
```
